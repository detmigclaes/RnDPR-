```python
w = np.linspace(0, 1*np.pi, 1000)

h = 0
for i in range(32):
    h += np.exp(1j*w)**(-i)

plt.figure()
plt.plot(w/np.pi, 20*np.log10(np.abs(h)))
h_cic = ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w)))
plt.plot(w/np.pi,20*np.log10(np.abs(h_cic)),label='cascade=1')
```

```python
# plot of cascaded integratpr-comb filter

# parameters
fs = 44100
R = 4
cascaded = 1

w = np.linspace(0, 1*np.pi, 1000)



#h_total = h_int*h_comb
h_cic = ((1-np.exp(-1j*w*R))/(1-np.exp(-1j*w)))**cascaded
R = 8
h_total1 = ((1-np.exp(-1j*w*R))/(1-np.exp(-1j*w)))**cascaded
R = 16
h_total2 = ((1-np.exp(-1j*w*R))/(1-np.exp(-1j*w)))**cascaded
plt.figure(figsize=(10, 5))
plt.plot(w/np.pi, np.abs(h_cic),label='R=4')
plt.plot(w/np.pi, np.abs(h_total1),label='R=8')
plt.plot(w/np.pi, np.abs(h_total2),label='R=16')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.legend()
#plt.yscale('log')
#plt.xscale('log')
plt.grid()
plt.savefig('figures/comb_filter.svg')
```

```python
# parameters
fs = 44100
R = 32
cascaded = 3

w = np.linspace(0, 1*np.pi, 10000)




R = 32
#h_cic = ((1-np.exp(-1j*w*R))/(1-np.exp(-1j*w)))**cascaded
#h_cic = ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w)))*((1-np.exp(-1j*w*16))/(1-np.exp(-1j*w)))*((1-np.exp(-1j*w*8))/(1-np.exp(-1j*w)))
#h_cic = ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w))) * ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w))) * ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w)))* ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w)))
h_cic = ((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w)))*((1-np.exp(-1j*w*32))/(1-np.exp(-1j*w))) * ((1-np.exp(-1j*w*16))/(1-np.exp(-1j*w))) *((1-np.exp(-1j*w*16))/(1-np.exp(-1j*w)))
# normalize to 0 dB at 0 Hz
h_cic /= h_cic[1]

#measure cutoff frequency

max_amplitude = max( 20 * np.log10(np.abs(h_cic[1:])))
threshold_db = max_amplitude - 3
threshold_amplitude = 10 ** (threshold_db / 20)

crossing_index = np.where(20*np.log10(np.abs(h_cic)) <= threshold_db)[0][0]
crossing_point = w[crossing_index]/np.pi#*fs/(2*np.pi)

peak_index = sp.signal.find_peaks(20*np.log10(np.abs(h_cic)))[0]
peak_index = np.append(crossing_index, peak_index)

# calculate the slope pr decade in the stopband

slope =(20*np.log10(np.abs(h_cic[peak_index[1]])) - 20*np.log10(np.abs(h_cic[peak_index[2]]))) / (np.log10(w[peak_index[1]]/np.pi) - np.log10(w[peak_index[2]]/np.pi))
plt.figure(figsize=(10, 5))


#plt.plot(w/np.pi, 20*np.log10(np.abs(h_cic)),label='SINC3 single downsamplers R=32' )
#plt.plot(w/np.pi, 20*np.log10(np.abs(h_cic)),label='SINC3 multible downsamplers R=32,16,8' )
#plt.plot(w/np.pi, 20*np.log10(np.abs(h_cic)),label='SINC4 single downsamplers R=32' )
plt.plot(w/np.pi, 20*np.log10(np.abs(h_cic)),label='SINC4 multiple downsamplers R=32,32,16,16' )

# plot cutoff frequency
#plt.plot([crossing_point, crossing_point], [0, 32], 'k--', label='cutoff frequency = {:.2f} $\pi$'.format(crossing_point))
plt.plot([0, 1], [threshold_db, threshold_db], 'k+-', label='3 dB cutoff frequency = {:.3f} $\pi$'.format(crossing_point))
plt.plot(w[peak_index]/np.pi, 20*np.log10(np.abs(h_cic[peak_index])), 'ro',)
plt.plot(w[peak_index]/np.pi, 20*np.log10(np.abs(h_cic[peak_index])), 'k--', label='slope = {:.2f} dB'.format(slope))
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')


plt.legend()
plt.ylim([-130,5])
#plt.ylim([25,30])
#plt.yscale('log')
plt.xscale('log')
plt.grid()


#plt.savefig('figures/SINC3_CIC_filter_R32.pdf')
#plt.savefig('figures/SINC3_CIC_filter_R32_16_8.pdf')
#plt.savefig('figures/SINC4_CIC_filter_R32.pdf')
plt.savefig('figures/SINC4_CIC_filter_R32_32_16_16.pdf')
```
